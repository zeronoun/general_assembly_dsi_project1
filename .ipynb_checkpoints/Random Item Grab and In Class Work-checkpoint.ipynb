```python
original_list = ["Dog", "Duckrack", "Mouse-Mouserat", "Housecat", "Catdog", "Wolf-wearwolf", "and-a-wearwolf"]
chosen_list = []
```

```python
from random import randint

def random_item():
    random_int = randint(0,len(original_list) - 1)
    random_item = original_list[random_int]
    return random_item

def random_replacement():
    current_random = random_item()
    
    while True:
        if current_random in chosen_list:
            current_random = random_item()
        elif current_random not in chosen_list:
            chosen_list.append(current_random)
            current_random = random_item()
        continue

    
    print current_random
    return chosen_list

random_replacement()
```
